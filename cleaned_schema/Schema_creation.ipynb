```python
import pandas as pd
import os
os.makedirs("cpsc368_group19/generated", exist_ok=True)
```

```python
tomatoes = pd.read_csv("cpsc368_group19/data/tomatoes_clean.csv")
oscars = pd.read_csv("cpsc368_group19/data/oscars_clean.csv")
letterboxd = pd.read_csv("cpsc368_group19/data/letterboxd_clean.csv")
```

```python
movies_df = tomatoes[['title', 'releaseDateTheaters', 'genre', 'originalLanguage']].copy()

movies_df = movies_df.rename(columns={
    'releaseDateTheaters': 'releaseDate',
    'originalLanguage': 'language'
})

movies_df.head()
```

```python
metrics_df = tomatoes[['title','audienceScore','tomatoMeter']].copy()
```

```python
metrics_df = metrics_df.merge(
    letterboxd[['movie_title','popularity','vote_average','vote_count']],
    left_on='title',
    right_on='movie_title',
    how='left'
)

metrics_df = metrics_df.drop(columns=['movie_title'])

metrics_df.head()
```

```python
oscars_df = oscars.copy()

oscars_df['title'] = oscars_df['film']

oscars_df = oscars_df[
    ['year_film','year_ceremony','ceremony','canon_category','name','title','winner']
]

oscars_df = oscars_df.reset_index(drop=True)

oscars_df['nomination_id'] = oscars_df.index + 1

oscars_df.head()
```

```python
movies_df.to_csv("cpsc368_group19/generated/movies_table.csv", index=False)
metrics_df.to_csv("cpsc368_group19/generated/movie_metrics_table.csv", index=False)
oscars_df.to_csv("cpsc368_group19/generated/oscar_nominations_table.csv", index=False)
```
